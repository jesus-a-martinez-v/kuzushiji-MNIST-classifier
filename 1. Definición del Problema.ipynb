¿Recuerdas cuando empezaste a programar? Es altamente probable que este haya sido tu primer programa:

### Conjuntos de Datos Kuzushiji

La comunidad, consciente de que la utilidad de nuestro buen amigo MNIST ha decaído gradualmente con el paso de los años, se ha abocado en la búsqueda de un digno substituto. Así, tenemos variantes como [Fashion MNIST](https://github.com/zalandoresearch/fashion-mnist) que toman la popular estructura del ya clásico conjunto de datos (imágenes de 28x28, en escala de grises, pertenecientes a 10 categorías disjuntas) y la aplican a un nuevo dominio, en este caso, prendas de ropa.

Otros optan por usar datos clásicos, pero a su vez más complejos, como [CIFAR-10](https://www.cs.toronto.edu/~kriz/cifar.html) o [SVHN](http://ufldl.stanford.edu/housenumbers/).

Una de las alternativas más interesantes, al menos en mi opinión, es [Kuzushiji-MNIST](https://github.com/rois-codh/kmnist) una alternativa considerablemente más difícil que el MNIST tradicional, puesto que se trata de reconocer caracteres japoneses escritos a mano alzada, en vez de números. Dichos caracteres presentan un mayor grado de variación que su contraparte numérica. 

Quizás lo más interesante de este dataset es que puede ayudar a reconstruir un lenguaje prácticamente perdido, Kuzushiji cursivo, el cual ha sido utilizado por casi 1000 años sin un estándar establecido e incluye docenas de estilos y formatos para la misma palabra. En el siglo XIX Japón reformó su lenguaje oficial, así como su sistema de escritura, estandarizando ambos, ocasionando que, con el tiempo, Kuzushiji cayera en desuso. En consecuencia, millones de documentos de la cultura e historia de Japón son inaccessibles para la mayoría de las personas.



En este proyecto exploraremos en mayor profundidad este conjunto de datos y, con suerte, podremos crear un modelo capaz de generar predicciones decentes sobre estos símbolos tan complejos e interesantes.
